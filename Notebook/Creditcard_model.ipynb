## Model Seletion, Training and Evaluation
We split the data into training and test sets, then train the selected models. The models are evaluated using accuracy, precision, F1-score, and recall.

* Storing Features Matrix in X and Target(response) in y

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score,f1_score,recall_score
```

* Splitting Dataset into Test and Train Dataset

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

Here are selecting <b>Decision Tree classifier</b>,<b>Logistic Regression </b> and Random Forest Classifier machine Models applying it on original dataset.

We are using here <b>Accuracy</b>, <b>Precision</b>, <b>Recall</b> and <b>F1 score</b> matrics to measure the performance of Machine Models.

```python
classsifier={"Logistic-Regression": LogisticRegression(),
            "Decision-Tree": DecisionTreeClassifier(),
            "Random Forest": RandomForestClassifier()}
for name, clf in classsifier.items():
    print(f"\n------{name}----")
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(f" Accuracy Score: {accuracy_score(y_test,y_pred)}")
    print(f" Precision: {precision_score(y_test,y_pred)}")
    print(f" F1-Score: {f1_score(y_test,y_pred)}")
    print(f" Recall: {recall_score(y_test,y_pred)}")
```

We got different score for each classfier as the the dataset is highly imbalance. Now we perform undersampling and oversampling and apply these classifier.

### <u>Undersampling</u>

```python
# classifying the leagit and Fraud transaction Dataset
normal=df[df['Class']==0]
fraud=df[df['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
#Undersampling the noraml transaction to 473 by deleting the data
normal_smple=normal.sample(n=473)
```

```python
normal_smple.shape
```

```python
Data=pd.concat([normal_smple,fraud], ignore_index=True)
Data
```

```python
Data['Class'].value_counts()
```

```python
X=Data.drop('Class',axis=1)
y=Data['Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
classsifier={"Logistic-Regression": LogisticRegression(),
            "Decision-Tree": DecisionTreeClassifier(),
            "Random Forest": RandomForestClassifier()
            }
for name, clf in classsifier.items():
    print(f"\n------{name}----")
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(f" Accuracy Score: {accuracy_score(y_test,y_pred)}")
    print(f" Precision: {precision_score(y_test,y_pred)}")
    print(f" F1-Score: {f1_score(y_test,y_pred)}")
    print(f" Recall: {recall_score(y_test,y_pred)}")
```

### <u>Oversampling</u>

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res= SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=.2,random_state=42)
```

```python
classsifier={"Logistic-Regression": LogisticRegression(),
            "Decision-Tree": DecisionTreeClassifier(),
            "Random Forest": RandomForestClassifier()}
for name, clf in classsifier.items():
    print(f"\n------{name}----")
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(f" Accuracy Score: {accuracy_score(y_test,y_pred)}")
    print(f" Precision: {precision_score(y_test,y_pred)}")
    print(f" F1-Score: {f1_score(y_test,y_pred)}")
    print(f" Recall: {recall_score(y_test,y_pred)}")
```

```python
# Train the model
RF=RandomForestClassifier()
RF.fit(X_res,y_res)
```
